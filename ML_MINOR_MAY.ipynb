```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from skimage.io import imread
from skimage.transform import resize

datadir = '/content/images'
CATEGORIES = ['pretty sunflower','rugby ball leather','ice cream cone'] # Name the folder with these categories
target = []
images = []
flat_data = []

for category in CATEGORIES:
  class_num=CATEGORIES.index(category)# label encoding the values
  path = os.path.join(datadir,category)# create path to use all the images
  #print(path)
  for img in os.listdir(path):
    img_array = imread(os.path.join(path,img))
    
    
    img_resize = resize(img_array,(150,150,3)) # Value gets converted 0-1 from 0-255
    flat_data.append(img_resize.flatten())
    images.append(img_resize)
    target.append(class_num)
    

flat_data = np.array(flat_data)
target = np.array(target)
images = np.array(images)
```

```python
df = pd.DataFrame(flat_data)
df['Target'] = target
df
```

```python
# split data into trianing and testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(flat_data,target,test_size=0.3,random_state=109)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn import svm
param_grid=[
            {'C':[1,10,100,1000],'kernel':['linear']},
            {'C':[1,10,100,1000],'gamma':[0.001,0.0001],'kernel':['rbf']}
                
            
]

svc=svm.SVC(probability=True)
clf=GridSearchCV(svc,param_grid)
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(y_pred,y_test)
```

```python
confusion_matrix(y_pred,y_test)
```
